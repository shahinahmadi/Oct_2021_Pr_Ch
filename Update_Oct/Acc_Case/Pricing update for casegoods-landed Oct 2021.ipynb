### Now we look into the method that we can append UFCA 1711E V4 file

```python
import numpy
from itertools import chain

CDN_OLD_CASE=pd.read_excel("APPENDING.xlsx",sheet_name="CASEGOODS V4")

Difference_SKUs=[]
INDEX=[]
for i in CDN_OLD_CASE["Option Item"]:
    if i not in CDN_SKUs:
          INDEX.append(CDN_OLD_CASE[CDN_OLD_CASE["Option Item"]==i].index.values)
          Difference_SKUs.append(i)
            
flatten_index = list(chain.from_iterable(INDEX))
print("This is the index of the numbers that are not in the new price list")
print(flatten_index)
print(Difference_SKUs)
```

### Now we look into the method of appending the old price list of UFCA 1711E V5

```python
import numpy
from itertools import chain

US_OLD_CASE=pd.read_excel("APPENDING.xlsx",sheet_name="CASEGOODS V5")

Difference_SKUs_US=[]
INDEX_US=[]
for i in US_OLD_CASE["Option Item"]:
    if i not in US_SKUs:
          INDEX_US.append(US_OLD_CASE[US_OLD_CASE["Option Item"]==i].index.values)
          Difference_SKUs_US.append(i)
            
flatten_index_US = list(chain.from_iterable(INDEX_US))
```

### We append 3 old lists to 3 new list as below 

```python
# Three lists for Canadian old price 
for i in flatten_index:
    PriceList.append("UFCA 1711E V5")
for i in flatten_index:
    Price.append(CDN_OLD_CASE.at[i,'Price'])
for i in Difference_SKUs:
    SKUs.append(i)
                 
# Three lists for US old price
for i in flatten_index_US:
    PriceList.append("UFMX 1711E V6")
for i in flatten_index_US:
    Price.append(US_OLD_CASE.at[i,'Price'])
for i in Difference_SKUs_US:
                 SKUs.append(i)
print(len(SKUs))
print(len(PriceList))
print(len(Price))
```

### Now we write the data to CSV file

```python
Table=[PriceList,SKUs,Price]
export_data = zip_longest(*Table, fillvalue = '')

with open('Upload_LANDED_ACCCASE_FINAL.csv', 'w', encoding="ISO-8859-1", newline='') as myfile:
      wr = csv.writer(myfile)
      wr.writerow(("PriceList", "ITEM", "Price"))
      wr.writerows(export_data)
myfile.close()
```
