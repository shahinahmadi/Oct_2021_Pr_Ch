```python
for i in CDN_Bedroom["CDN WHOLESALE (EX-WORKS)"]:
    Canadian_landed_Price.append(i)

for i in CDN_Entertainment["CDN WHOLESALE (EX-WORKS)"]:
    Canadian_landed_Price.append(i)
    
for i in CDN_Dining["CDN WHOLESALE (EX-WORKS)"]:
    Canadian_landed_Price.append(i)
    
for i in CDN_Office["CDN WHOLESALE (EX-WORKS)"]:
    Canadian_landed_Price.append(i)

for i in CDN_Storage["CDN WHOLESALE (EX-WORKS)"]:
    Canadian_landed_Price.append(i)

for i in CDN_ACC["CDN WHOLESALE (EX-WORKS)"]:
    Canadian_landed_Price.append(i)

for i in CDN_Custom["CDN WHOLESALE (EX-WORKS)"]:
    Canadian_landed_Price.append(i)

for i in CDN_Seasonal["CDN WHOLESALE (EX-WORKS)"]:
    Canadian_landed_Price.append(i)
    

    
    
for i in CDN_Bedroom["US WHOLESALE (EX-WORKS)"]:
    US_landed_Price.append(i)

for i in CDN_Entertainment["US WHOLESALE (EX-WORKS)"]:
    US_landed_Price.append(i)
    
for i in CDN_Dining["US WHOLESALE (EX-WORKS)"]:
    US_landed_Price.append(i)
    
for i in CDN_Office["US WHOLESALE (EX-WORKS)"]:
    US_landed_Price.append(i)

for i in CDN_Storage["US WHOLESALE (EX-WORKS)"]:
    US_landed_Price.append(i)

for i in CDN_ACC["US WHOLESALE (EX-WORKS)"]:
    US_landed_Price.append(i)

for i in CDN_Custom["US WHOLESALE (EX-WORKS)"]:
    US_landed_Price.append(i)

for i in CDN_Seasonal["US WHOLESALE (EX-WORKS)"]:
    US_landed_Price.append(i)


for i in Canadian_landed_Price:
    Price.append(i)
    
for i in US_landed_Price:
    Price.append(i)

print("*****************The length of the different columns in our data*************************")
print(len(PriceList))
print(len(SKUs))
print(len(Price))
print(Price)
```

```python
Table=[PriceList,SKUs,Price]
export_data = zip_longest(*Table, fillvalue = '')

with open('Upload_EXW_ACCCASE_FINAL.csv', 'w', encoding="ISO-8859-1", newline='') as myfile:
      wr = csv.writer(myfile)
      wr.writerow(("PriceList", "ITEM","Price"))
      wr.writerows(export_data)
myfile.close()
```
